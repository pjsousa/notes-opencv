```python
import notebook_util
import cv2
from matplotlib import pyplot as plt
import numpy as np
notebook_util.override_imshow(cv2)

%matplotlib inline
```

```python
image = cv2.imread('resources/car.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Create FAST detector object
#fast = cv2.FastFeatureDetector_create()
fast = cv2.xfeatures2d.StarDetector_create()
```

```python
# Create BRIEF extractor object
brief = cv2.xfeatures2d.BriefDescriptorExtractor_create()
```

```python
# Determine key points
keypoints = fast.detect(gray, None)
```

```python
# Obtain descriptors and new final keypoints using BRIEF
keypoints, descriptors = brief.compute(gray, keypoints)
print "Number of keypoints Detected: ", len(keypoints)
```

```python
# Draw rich keypoints on input image
cv2.drawKeypoints(image, keypoints, image, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
                                    
cv2.imshow('Feature Method - BRIEF', image)
```
