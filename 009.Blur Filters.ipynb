```python
import notebook_util
import cv2
from matplotlib import pyplot as plt
import numpy as np
notebook_util.override_imshow(cv2)

%matplotlib inline
```

```python
image_data = cv2.imread("./resources/cat.jpg")[180:400:, :]
cv2.imshow("Original Image", image_data, (15,5))
```

we can use the **cv2.blur** method and pass it the size for the kernel. cv2.blur performs a box filter with the kernel passed in.

```python
blur = cv2.blur(image_data, (3,3))
cv2.imshow('3x3 Kernel Blurring', blur, (15,5))
```

We can also use **cv2.filter2D** with a custom filter matrix

```python
kernel_3x3 = np.ones((3, 3), np.float32)
kernel_3x3
```

Although, when we define our filters "by hand" we need to make sure they are normalized (the sum of the filter values must be 1)

```python
## the kernels must always be normalized
kernel_3x3 = kernel_3x3 / 9
```

```python
blurred = cv2.filter2D(image_data, -1, kernel_3x3)
cv2.imshow('3x3 Kernel Blurring', blurred, (15,5))
```

```python
kernel_7x7 = np.ones((7, 7), np.float32)
kernel_7x7 = kernel_7x7 / np.sum(kernel_7x7)
```

```python
blurred = cv2.filter2D(image_data, -1, kernel_7x7)
cv2.imshow('3x3 Kernel Blurring', blurred, (15,5))
```

We can also use a gausian kernel with **cv2.GaussianBlur**

```python
Gaussian = cv2.GaussianBlur(image_data, (7,7), 0)
cv2.imshow('Gaussian Blurring', Gaussian, (15,5))
```

**cv2.medianBlur**

```python
median = cv2.medianBlur(image_data, 5)
cv2.imshow('Median Blurring', median, (15,5))
```

**cv2.bilateralFilter**

```python
bilateral = cv2.bilateralFilter(image_data, 9, 75, 75)
cv2.imshow('Bilateral Blurring', bilateral, (15,5))
```
