
OpenCV retrieves the data in BGR format. This means we do have three channels. And they to correspond to the usual Red, Green and Blue colors.

But we have been misguided by our imshow function.

```python
import matplotlib.pyplot as plt
```

If we use pyplot to look again at the car image, the reds from the buttons panels and the display turn blue. This just means that where pyplot is especting to see values corresponding to "blue" (the last position) we actually have the "red" values.

```python
plt.imshow(car_data, aspect="auto")
```

If we do the same with our cake image. We can notice that the red bowtie appears blue. But the green cake stays the same (the green channel is in the expected position)

```python
plt.imshow(cake_data, aspect="auto")
```

Lets look at an individual channel.

```python
# Since we're using opencv raw data we have:
# 3 channels > [BLUE, GREEN, RED]
cv2.imshow("Blue channel (we don't have too noticable blue color in the image)", cake_data[:,:,0], (7,5))
cv2.imshow("Green channel (the green cake 'ligths up' in this channel compared to the other two)", cake_data[:,:,1], (7,5))
cv2.imshow("Red channel (the bowtie 'lights up' and almost mixes with the actual white color of the sheep)", cake_data[:,:,2], (7,5))
```

**Re-ordering channels**

We can reorder channels by doing:

```python
bgr_image = cake_data
rgb_image = cv2.cvtColor(bgr_image,cv2.COLOR_BGR2RGB)

print("The original raw BGR image:")
plt.imshow(bgr_image)
plt.show()

print("The RGB image after cv2.cvtColor:")
plt.imshow(rgb_image)
plt.show()
```
