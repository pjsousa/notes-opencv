```python
import notebook_util
import cv2
from matplotlib import pyplot as plt
import numpy as np
notebook_util.override_imshow(cv2)

%matplotlib inline
```

```python
image_data = cv2.imread('./resources/hello.png')
_, image_data = cv2.threshold(image_data, 0, 255, cv2.THRESH_BINARY_INV)
cv2.imshow("Original", image_data)
```

```python
# Let's define our kernel size
kernel = np.ones((5,5), np.uint8)

# Now we erode
erosion = cv2.erode(image_data, kernel, iterations = 1)
cv2.imshow('Erosion', erosion)
#cv2.waitKey(0)

```

```python
dilation = cv2.dilate(image_data, kernel, iterations = 1)
cv2.imshow('Dilation', dilation)
```

```python
# Opening - Good for removing noise
opening = cv2.morphologyEx(image_data, cv2.MORPH_OPEN, kernel)
cv2.imshow('Opening', opening)
#cv2.waitKey(0)

```

```python
# Closing - Good for removing noise
closing = cv2.morphologyEx(image_data, cv2.MORPH_CLOSE, kernel)
cv2.imshow('Closing', closing)

```
