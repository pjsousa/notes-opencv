```python
import notebook_util
import cv2
from matplotlib import pyplot as plt
import numpy as np
notebook_util.override_imshow(cv2)

%matplotlib inline
```

```python
image_data = cv2.imread("./resources/cake2.jpg")
image_data = image_data[250:500,400:]
```

```python
cv2.imshow("Original Image", image_data, (15,7))
```

```python
dst = cv2.fastNlMeansDenoisingColored(image_data, None, 6, 6, 7, 21)

cv2.imshow('Fast Means Denoising', dst, (15,7))
```

OpenCV as multiple Non-Local Means denoising implementations:
- cv2.fastNlMeansDenoising() - works with a single grayscale images
- cv2.fastNlMeansDenoisingColored() - works with a color image.
- cv2.fastNlMeansDenoisingMulti() - works with image sequence captured in short period of time (grayscale images)
- cv2.fastNlMeansDenoisingColoredMulti() - same as above, but for color images.
