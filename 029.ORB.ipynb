```python
import notebook_util
import cv2
from matplotlib import pyplot as plt
import numpy as np
notebook_util.override_imshow(cv2)

%matplotlib inline
```

```python
image = cv2.imread('resources/car.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Create ORB object, we can specify the number of key points we desire
orb = cv2.ORB_create()
```

```python
# Determine key points
keypoints = orb.detect(gray, None)
```

```python
# Obtain the descriptors
keypoints, descriptors = orb.compute(gray, keypoints)
print("Number of keypoints Detected: ", len(keypoints))
```

```python
# Draw rich keypoints on input image
cv2.drawKeypoints(image, keypoints, image, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
cv2.imshow('Feature Method - ORB', image)
```
