## Draw Polygons

```python
all_zeros_rgb = np.zeros((512,512,3), np.uint8)

# Let's define four points
# these are 4 points in image pixel coordinates
pts = np.array([
    [400,50], 
    [130,200],
    [10,400],
    [145,500]
    ], np.int32)

print("pts needs to go from this:\n {}".format(pts))
print("shape: {}".format(pts.shape))

# We need to reshape our points in order to 
# be compliant with cv2.polyline
pts = pts.reshape((-1,1,2))

print("\nto this:\n {}".format(pts))
print("shape: {}".format(pts.shape))

cv2.polylines(all_zeros_rgb, [pts], True, (255,127,0), 5)
cv2.imshow("Polygon", all_zeros_rgb)
```

**drawing polygons, the points order matters**

```python
all_zeros_rgb = np.zeros((512,512,3), np.uint8)

#these are same coordinates but with a different order
pts = np.array([
    [400,50], 
    [130,200],
    [145,500],
    [10,400]
    ], np.int32)

pts = pts.reshape((-1,1,2))

cv2.polylines(all_zeros_rgb, [pts], True, (255,127,0), 5)
cv2.imshow("Polygon", all_zeros_rgb)
```

## Draw text

**cv2.putText(image, 'Text to Display', bottom left starting point, Font, Font Size, Color, Thickness)**

- FONT_HERSHEY_SIMPLEX
- FONT_HERSHEY_PLAIN
- FONT_HERSHEY_DUPLEX
- FONT_HERSHEY_COMPLEX 
- FONT_HERSHEY_TRIPLEX
- FONT_HERSHEY_COMPLEX_SMALL
- FONT_HERSHEY_SCRIPT_SIMPLEX
- FONT_HERSHEY_SCRIPT_COMPLEX

```python
all_zeros_rgb = np.zeros((512,512,3), np.uint8)

coords_pts = np.array([
    [400,50], 
    [130,200],
    [145,500],
    [10,400]
    ], np.int32)

pts = coords_pts.reshape((-1,1,2))

cv2.polylines(all_zeros_rgb, [pts], True, (255,127,0), 5)

for i, (x,y) in enumerate(coords_pts):
    cv2.putText(all_zeros_rgb, str(i+1), (x,y), cv2.FONT_HERSHEY_COMPLEX, 1, (0,255,240), 2)

cv2.imshow("Hello World!", all_zeros_rgb)
```
